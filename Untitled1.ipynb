```python
import numpy as np
import pandas as pd
name=pd.read_csv("C:\\Users\\HP\\Desktop\\Main File\\Future Engineering campus x\\Social_Network_Ads.csv")
name
```

```python
name.head()
```

```python
name=name.iloc[:,3:]
name
```

```python
name.shape
```

```python
# Train _Test Split karna
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(name.drop('Purchased',axis=1),
                                                         name['Purchased'],
                                                         test_size=0.3,
                                                         random_state=0)

x_train.shape,x_test.shape                                    
```

```python
# Data par satandardizication karne par

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

# Data Train ke liye Send karna Fit Only Train Data ko hi kare
scaler.fit(x_train)


# Tranform  Dono Data Train_test ko kare
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
scaler.mean_
```

```python
# scaled Data ko DataFrame me conver karna

x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
# value ko Roundup karna
np.round(x_train.describe(),1)
```

```python
np.round(x_train_scaled.describe(),1)
```
