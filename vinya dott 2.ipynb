```python
import numpy as np 
import pandas as pd 
from sklearn.compose import ColumnTransformer 
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("C:\\Users\\HP\\Desktop\\Projects\\Project Data\\covid_toy.csv")
df
```

```python
x = df.drop(columns = ['has_covid'])
y = df['has_covid']
```

```python
from sklearn.model_selection import train_test_split 
x_train , x_test , y_train ,y_test = train_test_split(x,y,test_size=  0.2 , random_state = 42)
```

```python
transformer = ColumnTransformer(transformers = [
    ('a' , SimpleImputer() , ['fever']) ,  
    ('b' , OrdinalEncoder(categories = [['Mild' , 'Strong']]) , ['cough']),
    ('c' , OneHotEncoder(sparse = False , drop = 'first') , ['gender' , 'city']),
    ("e"  , LabelEncoder(),["has_covid"])
] , remainder = 'passthrough')
```

```python
cff = transformer.fit_transform(x_train)
```

```python
cff = pd.DataFrame(cff)
```
